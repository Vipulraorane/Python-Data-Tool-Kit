# Restful API & Flask — Answer Book


## Question
What is a RESTful API?


## Answer
A RESTful API is an application programming interface that follows REST (Representational State Transfer) principles. It allows communication between client and server using HTTP methods and resources identified by URLs.


## Question
What is Flask, and why is it popular for building APIs?


## Answer
Flask is a lightweight Python web framework used to build web applications and APIs. It is popular because it is simple, flexible, easy to learn, and requires minimal setup compared to larger frameworks.


## Question
What are HTTP methods used in RESTful APIs?


## Answer
Common HTTP methods used in RESTful APIs include:
- GET: Retrieve data
- POST: Create new data
- PUT: Update existing data
- PATCH: Partially update data
- DELETE: Remove data


## Question
What is the purpose of the @app.route() decorator in Flask?


## Answer
The @app.route() decorator in Flask is used to map a URL endpoint to a Python function. When a user accesses that URL, the associated function executes and returns a response.


## Question
What is the role of Flask-SQLAlchemy?


## Answer
Flask-SQLAlchemy is an extension that integrates SQLAlchemy with Flask. It simplifies database operations by providing ORM (Object Relational Mapping) support for working with databases using Python classes.


## Question
How do you create a basic Flask application?


## Answer
A basic Flask application is created by importing Flask, creating an app instance, defining routes, and running the app.


## Question
How do you return JSON responses in Flask?


## Answer
JSON responses in Flask can be returned using the jsonify() function, which converts Python dictionaries into JSON format.


## Question
How do you handle POST requests in Flask?


## Answer
POST requests in Flask are handled by specifying methods=['POST'] in the route decorator and accessing data using request.form or request.json.


## Question
How do you handle errors in Flask (e.g., 404)?


## Answer
Errors in Flask can be handled using error handlers with @app.errorhandler(). For example, a 404 error handler returns a custom response when a page is not found.


## Question
How do you structure a Flask app using Blueprints?


## Answer
Flask Blueprints are used to organize large applications into smaller modules. They help separate routes, templates, and logic into reusable components.
